```python
import numpy as np
import pandas as pd
```

PM1

```python
df = pd.read_excel("Dsata Set for Assignment 1.xlsx")
df = df.dropna()
df['diagnosis'] = df['diagnosis'].replace({"B": -1, "M": 1})
df_train = df.sample(frac=0.67, random_state=59)
df_test = df.drop(df_train.index)
X_train = df_train.loc[:, "radius_mean":]
Y_train = df_train.loc[:, "diagnosis":"diagnosis"]
X_test = df_test.loc[:, "radius_mean":]
Y_test = df_test.loc[:, "diagnosis":"diagnosis"]
W = X_train.iloc[222].copy()
W[:] = 0
num_rows = df_train.shape[0]
remaining_rows = df.shape[0] - num_rows
epochs = 100
for epoch in range(epochs):
    for i in range(num_rows):
        req = W.mul(X_train.iloc[i], fill_value=1)
        ans = req.sum()
        Y_pred = 1 if ans >= 0 else -1
        y_i = Y_train.values[i]
        if(y_i != Y_pred):
            W += X_train.iloc[i] * y_i


def test(k):
    check = W.mul(X_test.iloc[k], fill_value=1)
    test = check.sum()
    return 1 if test >= 0 else -1


def checkaccuracy():
    count = 0
    for i in range(remaining_rows):
        value = test(i)
        y_test = Y_test.values[i]
        if(value == y_test):
            count += 1
    accuracy = count*100
    return accuracy/remaining_rows


def checkprecision():
    true_positives = 0
    total_positives = 0
    for i in range(remaining_rows):
        value = test(i)
        y_test = Y_test.values[i]
        if(value == 1):
            total_positives += 1
            if(y_test == 1):
                true_positives += 1
    precision = true_positives/total_positives
    return precision*100


def checkrecall():
    true_positives = 0
    false_negatives = 0
    for i in range(remaining_rows):
        value = test(i)
        y_test = Y_test.values[i]
        if(value == 1):
            if(y_test == 1):
                true_positives += 1
        else:
            if(y_test == 1):
                false_negatives += 1
    recall = true_positives/(true_positives+false_negatives)
    return recall*100


accuracy = checkaccuracy()
precision = checkprecision()
recall = checkrecall()

print(accuracy)
print(precision)
print(recall)
```
